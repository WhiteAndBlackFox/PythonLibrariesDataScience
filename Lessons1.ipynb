# Урок 1. Введение в задачу классификации. Постановка задачи и подготовка данных.

## 1. Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?

**Precision (точность)** - показывает, сколько всего результатов было предсказано верно. Другими словами, точность дает нам ответ на вопрос «Из всех объектов, которые классифицированы как принадлежащие классу, сколько на самом деле принадлежит ему?». 

Приоритет на данную метрику необходим в том случае, когда критическим является ложное срабатывание, но нужжно учитывать, чем больше точность, тем меньше положительных попаданий в *матрице ошибок* (**positive predictive value** - PPV). Данный инцидент может сказаться на предсказании связанные с финнансывыми затратами, из-за чего каждое дополнительное ложное попадание черевато нецелесообразными дополнительными финансовыми затратами или потерями для бизнеса.

В качестве примера можно привести:

* обработка фото (фото должно должны обрабатываться в высокой точность);
* выдача кредита (выгодно выдавать благонадёжным людям, при не является критической ошибкой в отказе какому то проценту людей).

Следовательно, повышение метрики Precision позволяет бизнесу оптимизировать статьи расходов и увеличивать статьи доходности.

**Recall (полнота)** - показывает, сколько истинных результатов было предсказано верно. То есть, полнота показывает сколько образцов из всех положительных примеров были классифицированы правильно. Чем выше значение полноты, тем меньше положительных примеров пропущено в классификации.

Так как данная метрика характеризует способность классификатора определять, как можно большее число положительных ответов из ожидаемых, отдавать её приоритет необходимо в тех случаях когда нужно, как можно меньше пропустить целевых объектов, а в случае ошибочного отнесения направить на дополнительное обследование.

В качестве примера можно привести:

* важно выявить всех несоответвствующих требованиям претендентов на должность или участие в проекте, доступу к информации, при этом возможно включив в это число претендентов соответствующих требованиям, предоставив им возможность повторного участия в проверке на соответствие;
* выявление брака на производстве, где отбраковку можно направить ручную проверку;
* важно выявить всех пациентов больных каким-либо заболеванием, при этом возможно включив в это число здоровых пациентов и позже отправив их на дообследование.

Следовательно, повышение метрики Recall позволяет бизнесу сократить риски и оптимизировать бизнесс-процессы.

## 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?


**F-мера** – взвешенное **гармоническое среднее** полноты и точности. Этот показатель демонстрирует, как много случаев прогнозируется моделью правильно, и сколько истинных экземпляров модель не пропустит. F-мера объединяет в себе информацию о полноте и точности используемой модели

У гармонического среднего есть свойство: если один из членов ряда стремится к нулю, то и гармоническое среднее стремится к нулю, т.е. таким образом мы можем выявить, что один из показателей имеет недостаточную величину. В случае с простым арифметическим средним один показатель может компенсироваться за счет другого.

Всегда необходимо находить компромисс между оптимизацией полноты и оптимизацией точности, и делать это исходя из конкретной предметной области в поле которой решается определённая задача классификации. Т.е. в некоторых случаях будет важнее и существенней влияние на верный конечный результат метрики *Precision*, а в других случаях будет приоритетней влияние метрики *Recall*. Именно поэтому мы не можем просто взять среднее от этих метрик, поскольку преобладающее влияние одной из них на результат классификации принципиально важно и определяет его в конечном итоге.
