# Урок 2. Анализ данных и проверка статистических гипотез.

## 1. В чём различие между зависимыми и независимыми выборками?

Выборки называются **зависимыми**, когда речь идет об одной группе испытуемых до и после работы с данными, т.е. когда можно установить взаимооднозначную связь между случаями в выборках.

Выборки называются **независимыми**, когда в работе используются разные группы, т.е. когда связь между случаями установить невозможно.


## 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Все критерии различий условно подразделены на две группы: **параметрические** и **непараметрические** критерии.

Критерий различия называют параметрическим, если в группе статистических критериев, включаются в расчет параметры вероятностного распределения признака.

Параметрические критерии, применяются при известном законе распредления генеральной совокупности, как правило нормальном или при использовании параметров данной совокупности (среднее, дисперсия). При нормальном распределении генеральной совокупности параметрические критерии обладают большей мощностью по сравнению с непараметрическими. Иными словами, они способны с большей достоверностью отвергать нулевую гипотезу, если последняя неверна.

Критерий различия называют непараметрическим, когда группа статистических критериев, которые не включают в расчёт параметры вероятностного распределения и основаны на оперировании частотами или рангами.

На практике, большинство изучаемых данных, не подлежат номарльному закону распределния. Применение в таких случаях параметрических критериев может привести к ошибкам в статистических выводах. В таких случаях непараметрические критерии оказываются более мощными, т.е. способными с большей достоверностью отвергать нулевую гипотезу.
